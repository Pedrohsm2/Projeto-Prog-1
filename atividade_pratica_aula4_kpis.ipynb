## 7. Análise por canal

### Pergunta de negócio
**Qual canal de venda é o mais eficiente?**

### Tarefa
Crie uma tabela agregada por `canal_venda` contendo, no mínimo:
- Receita Total
- Lucro Total
- Margem Média ou Margem Calculada
- Quantidade de pedidos
- Ticket Médio

### Depois responda
1. Qual canal gera mais receita?
2. Qual canal gera maior eficiência?
3. Se você fosse gestor, priorizaria volume ou margem?


```python
tabela_canal = df.groupby('canal_venda').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'canal_venda': 'count'
})

tabela_canal.columns = ['receita_total', 'lucro_total', 'qtd_pedidos']

# margem
tabela_canal['margem'] = tabela_canal['lucro_total'] / tabela_canal['receita_total']

# ticket médio
tabela_canal['ticket_medio'] = tabela_canal['receita_total'] / tabela_canal['qtd_pedidos']

tabela_canal
```

1. Qual canal gera mais receita? **R: Marketplace**
2. Qual canal gera maior eficiência? **R: Marketplace (melhor margem também)**
3. Você priorizaria volume ou margem? **R: Volume pois a marrgem da loja física está bem proxima do marketplace.**

### Insight obrigatório
Escreva 1 ou 2 frases interpretando sua tabela, no estilo executivo:
- não apenas “o Online foi maior”
- mas o que isso **significa para a decisão**


**O canal Marketplace lidera tanto em receita quanto em eficiência, sendo o principal responsável pelo desempenho do negócio. Como a margem da loja física é próxima, priorizar volume (escala) faz mais sentido, focando nos canais que geram maior receita.**

## 8. Análise por categoria

### Pergunta de negócio
**Qual categoria é mais lucrativa e qual parece mais eficiente?**

### Tarefa

1. A categoria de maior receita também é a de maior margem?

```python
tabela_categoria = df.groupby('categoria').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'categoria': 'count'
})

tabela_categoria.columns = ['receita_total', 'lucro_total', 'qtd_pedidos']

# margem
tabela_categoria['margem'] = tabela_categoria['lucro_total'] / tabela_categoria['receita_total']

# ticket médio
tabela_categoria['ticket_medio'] = tabela_categoria['receita_total'] / tabela_categoria['qtd_pedidos']

tabela_categoria
```

2. Existe categoria que vende bem mas é pouco eficiente? **R: Sim, Informática e Telefonia.**

## 9. Análise por região (UF)

### Pergunta de negócio
**Quais UFs concentram a operação?**

### Tarefa
Monte uma tabela por `uf` com:
- Receita Total
- Lucro Total
- Participação percentual da receita
- Ticket Médio

### Depois responda
1. Quais UFs concentram maior receita?
2. Existe concentração regional?
3. Se fosse necessário expandir ou reforçar logística, por onde você começaria?


```python
tabela_uf = df.groupby('uf').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'uf': 'count'
})

tabela_uf.columns = ['receita_total', 'lucro_total', 'qtd_pedidos']

# participação da receita
tabela_uf['participacao'] = tabela_uf['receita_total'] / tabela_uf['receita_total'].sum()

# ticket médio
tabela_uf['ticket_medio'] = tabela_uf['receita_total'] / tabela_uf['qtd_pedidos']

tabela_uf
```

1. Quais UFs concentram maior receita? **R: SP e RJ.**
2. Existe concentração regional? **R: Não, está bem distribuído.**
3. Onde começar expansão/logística? **R: Pelas UFs que tem maior receita para que se tenha um retorno mais rápido e entao para as UFs que tem menor receita para buscar uma oportunidade de mercado.**
